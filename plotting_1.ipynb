```python
import json
from utils import plot_line_graph

# read estimated_params.json
with open('estimated_params.json', 'r') as f:
    estimated_params = json.load(f)

bics = []
logliks = []

for key, value in estimated_params.items():
    bics.append(value['BIC'])
    logliks.append(value['log_likelihood'])

plot_line_graph(bics, "BIC")
plot_line_graph(logliks, "Log-Likelihood")
```
