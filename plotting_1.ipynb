```python
import json
import pandas as pd
from utils import plot_line_graph, insert_intercept, plot_probabilities

# read estimated_params.json
with open('estimated_params_546689.json', 'r') as f:
    estimated_params = json.load(f)

data = pd.read_csv("562606.csv")

X = data.Price.values
X = insert_intercept(X.reshape(-1, 1)) # Creates a T by 2 design matrix

bics = []
logliks = []

for key, value in estimated_params.items():
    bics.append(value['BIC'])
    logliks.append(value['log_likelihood'])

plot_line_graph(bics, "BIC")
plot_line_graph(logliks, "Log-Likelihood")

print(estimated_params)

plot_probabilities(estimated_params, K=3, X=X)
```
